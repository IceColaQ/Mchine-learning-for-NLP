```python
from collections import defaultdict;
freqs = defaultdict(Counter)
with open('a1_data/books.txt', encoding='ISO-8859-1') as f:
    for line in f:
    	tokens = line.lower().split()
    	for t1, t2 in zip(tokens, tokens[1:]):
    	    freqs[t1][t2] += 1
freqs['red'].most_common(1)
```
